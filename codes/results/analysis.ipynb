```python
import pandas as pd

# -------------------------------
# Load the data
# -------------------------------
input_file = "results_deterministic_approximate.csv"
df = pd.read_csv(input_file, header=None, names=["file", "objective", "time", "solutions"])

# -------------------------------
# Robust filename parser
# -------------------------------
def split_filename(fname):
    fname = fname.replace(".json", "")
    parts = fname.split("_")
    
    # Handle short/malformed names
    if len(parts) < 4:
        return pd.Series([fname, "", "", ""])  # fallback: just filename

    # Multi-part instance names (e.g., transoptas_5pso, deepela_large_r1)
    if len(parts) > 4:
        instance_name = "_".join(parts[:-3])
        measure = parts[-3]
        percentile = parts[-2]
        algorithm = parts[-1]
    else:
        instance_name = parts[0]
        measure = parts[1]
        percentile = parts[2]
        algorithm = parts[3]

    return pd.Series([instance_name, measure, percentile, algorithm])

# Apply parser
df[["instance_name", "measure", "percentile", "algorithm"]] = df["file"].apply(split_filename)

# Reorder columns
df = df[["instance_name", "measure", "percentile", "algorithm", "objective", "time", "solutions"]]

# Preview
df.head(20)
```

```python
#results_kmis_online_mis.csv


import pandas as pd
import ast

# Load CSV
nclue_file = "results_kmis_online_mis.csv"

df_nclue = pd.read_csv(nclue_file)
df_nclue.columns = df_nclue.columns.str.strip()


def split_nclue_filename(fname):
    fname = fname.replace(".out", "")
    parts = fname.split("_")
    
    # The last three parts are measure, percentile, seed
    seed = parts[-1]
    percentile = parts[-2]
    measure = parts[-3]
    
    # Everything before that is instance name
    instance_name = "_".join(parts[:-3])
    
    return pd.Series([instance_name, measure, percentile, seed])

# Apply parser
df_nclue[["instance_name", "measure", "percentile", "seed"]] = df_nclue["file"].apply(split_nclue_filename)

# Convert solution string to list if needed
df_nclue["solution"] = df_nclue["solution"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Reorder columns
df_nclue = df_nclue[["instance_name", "measure", "percentile", "seed", "solution", "objective"]]

# Preview
df_nclue.head(20)
```

```python
# results_redumis.csv:

import pandas as pd
import ast

# Load your CSV/TSV
df_redumis = pd.read_csv("results_redumis.csv")  # or sep="," / sep="\t" depending on file
df_redumis.columns = df_redumis.columns.str.strip()


# -------------------------------
# Step 1: Correct parser
# -------------------------------
def split_redumis_filename(fname):
    fname = fname.replace(".out", "")
    parts = fname.split("_")
    
    # Identify known measures
    measure_candidates = ["correlation", "cosine", "seuclidean"]
    measure = next((m for m in parts if m in measure_candidates), "")
    
    # Instance name: everything before measure
    measure_idx = parts.index(measure) if measure else 0
    instance_name = "_".join(parts[:measure_idx])
    
    # Percentile: first part that starts with 'p'
    percentile = next((p for p in parts if p.startswith("p")), "")
    
    # Seed: if "seed" is present
    seed = ""
    if "seed" in parts:
        seed_idx = parts.index("seed")
        seed = parts[seed_idx + 1]
    
    return pd.Series([instance_name, measure, percentile, seed])

# -------------------------------
# Step 2: Apply parser
# -------------------------------
df_redumis[["instance_name", "measure", "percentile", "seed"]] = df_redumis["file"].apply(split_redumis_filename)

# -------------------------------
# Step 3: Convert solution string to Python list
# -------------------------------
df_redumis["solution"] = df_redumis["solution"].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith("[") else []
)

# -------------------------------
# Step 4: Reorder columns
# -------------------------------
df_redumis = df_redumis[["instance_name", "measure", "percentile", "seed", "solution", "objective"]]

# -------------------------------
# Step 5: Preview
# -------------------------------
df_redumis.head(20)
```
